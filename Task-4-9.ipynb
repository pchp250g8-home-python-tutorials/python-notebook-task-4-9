Из 40 тонн руды выплавили 20 тонн металла, содержащего 6% примесей.
Каков процент примесей в руде?

```python
from sympy.abc import x
from sympy import Eq
from sympy import solve

mm1 = 20
mr = 40
pg = 6
mm2 = mm1 - mm1 * pg  / 100
ans = (mr - mm2) / mr * 100
print(ans)
```

```python
from sympy.abc import x, y
from sympy import Eq
from sympy import solve

sol_eq = solve([Eq((20 - x) / 20 * 100, 6)], 
                [x])
print(sol_eq)
ans = (40 - sol_eq[x]) / 40 * 100
print(ans)
```
